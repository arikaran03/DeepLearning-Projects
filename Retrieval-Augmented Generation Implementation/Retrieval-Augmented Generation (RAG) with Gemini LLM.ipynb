```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
```

```python
#example sentence
sentences = [
    "I love cates",
    "Dogs are best",
    "Artificial intelligence is the feture",
    "Cats are cute animals",
    "I enjoy machine learning",
    "I have a pet dog"
]
```

```python
# Load pre-traind sentence embedding model
gemini_api_key = "YOUR_GEMINI_API_KEY"
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(sentences)  # shape (6,384) 6 values, 384 Dimensional vector
embeddings = np.array(embeddings).astype('float32')
embeddings.shape
```

```python
dimension = embeddings.shape[1] #picking 364 dim value
index = faiss.IndexFlatL2(dimension) # here we find indices are all the similar words using Euclidian distance formula
index.add(embeddings)
```

```python
query = "I adore kittens"
query_vecotor = model.encode([query]).astype('float32')
```

```python
top_k = 3
distances, indices = index.search(query_vecotor,top_k)
```

```python
# indices.shape
for i in indices[0]: # here 0 represents for the number of queries that we are given to the model we given only 1 so the index is 0
  print(sentences[i])
```
